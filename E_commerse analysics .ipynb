```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
import seaborn as sns
```

```python
df = pd.read_csv("Ecommerce_Consumer_Behavior_Analysis_Data.csv")
pd.set_option('display.max_columns', None)
df.head(70)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
mode_value = df["Social_Media_Influence"].mode()[0]
df["Social_Media_Influence"].fillna(mode_value, inplace=True)
df.head(6)
```

```python
mode_value = df["Engagement_with_Ads"].mode()[0]
df["Engagement_with_Ads"].fillna(mode_value, inplace=True)
df.head(5)
```

```python
plt.figure(figsize=(9,5))
sns.countplot(x="Gender",data=df,color = "red")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.hist(df["Age"], bins=10, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x="Income_Level",data=df,color = "red",width = 0.3)
plt.show()
```

```python
df["Purchase_Amount"] = df["Purchase_Amount"].astype(str)
df["Purchase_Amount"] = df["Purchase_Amount"].str.replace("$", "", regex=False)
df["Purchase_Amount"] = df["Purchase_Amount"].str.replace(",", "", regex=False)
df["Purchase_Amount"] = df["Purchase_Amount"].str.strip()
df["Purchase_Amount"] = pd.to_numeric(df["Purchase_Amount"], errors="coerce")
print(df["Purchase_Amount"].dtype)
print(df["Purchase_Amount"].head(10))
```

```python
bins = [18, 25, 35, 45, 50]   
labels = ["18-24", "25-34", "35-44", "45-50"]
df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)
sns.barplot(x="Age_Group", y="Purchase_Amount", data=df, estimator="mean",width=0.3)
plt.title("Age Group vs Average Purchase Amount")
plt.xlabel("Age Group")
plt.ylabel("Average Purchase Amount")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Income_Level", y="Frequency_of_Purchase", data=df, estimator="mean",width=0.3)
plt.title("Income Level vs Purchase Frequency")
plt.xlabel("Income Level")
plt.ylabel("Average Purchase Frequency")
plt.show()
```

```python
df["Purchase_Amount"] = df["Purchase_Amount"].astype(str)
df["Purchase_Amount"] = df["Purchase_Amount"].str.replace("$", "", regex=False)
df["Purchase_Amount"] = df["Purchase_Amount"].str.replace(",", "", regex=False)
df["Purchase_Amount"] = df["Purchase_Amount"].str.strip()
df["Purchase_Amount"] = pd.to_numeric(df["Purchase_Amount"], errors="coerce")
print(df["Purchase_Amount"].dtype)
print(df["Purchase_Amount"].head(10))
```

```python
df["OrderDate"] = pd.to_datetime(df["Time_of_Purchase"], errors="coerce")
df["Month"] = df["OrderDate"].dt.to_period("M")
month_sales = df.groupby("Month")["Purchase_Amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(9,5))
month_sales.plot(kind="bar", color="skyblue")
plt.title(" Month-wise Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Week"] = df["OrderDate"].dt.isocalendar().week  # week number
week_sales = df.groupby("Week")["Purchase_Amount"].sum()

plt.figure(figsize=(12,5))
week_sales.plot(kind="bar", color="green")
plt.title("Week-wise Sales (Week Number 1-52)")
plt.xlabel("Week Number")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```
