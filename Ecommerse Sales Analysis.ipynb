Dataset Overview (Short Version)
This dataset contains sales transaction data used to analyze customer behavior and product performance. It includes columns such as Date, Customer Age, Gender, Country, Product Category, Quantity, Cost, and Revenue.
The dataset helps identify sales trends, top-selling products, and revenue patterns across different locations and customer groups. 📊

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop("Column1", axis=1)
df
```

```python
# Numeric columns
numeric_cols = ["Year", "Customer Age", "Quantity", "Unit Cost", "Unit Price", "Cost", "Revenue"]
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean()) 
```

```python
df.isnull().sum()
```

```python
categorical_cols = ["Month", "Customer Gender", "Country", "State", "Product Category", "Sub Category"]
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0]).astype("category")

```

```python
df.isnull().sum()
```

```python
df["Date"] = df["Date"].ffill()
```

```python
df.isnull().sum()
```

```python
# Integer conversion
df["Year"] = df["Year"].astype("int")
df["Customer Age"] = df["Customer Age"].astype("int")
df["Quantity"] = df["Quantity"].astype("int")
```

```python
# Categorical conversion
categorical_cols = ["Month", "Customer Gender", "Country", "State", "Product Category", "Sub Category"]
for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
df.dtypes
```

```python
df.head()
```

```python
monthly_sales = df.groupby("Month", observed=True)["Revenue"].sum()

plt.figure(figsize=(6,4))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)

plt.show()
```

```python


category_revenue = df.groupby("Product Category")["Revenue"].sum().reset_index()
plt.figure(figsize=(5,3))
sns.barplot(x="Product Category", y="Revenue", data=category_revenue, color="skyblue",width=0.4)
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

3️⃣ Best Selling Sub Categories 📦

```python

sub_sales = df.groupby("Sub Category", observed=True)["Quantity"].sum().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(
    data=sub_sales,
    x="Sub Category",
    y="Quantity",
    hue="Sub Category",
    legend=False
)
plt.title("Best Selling Sub Categories")
plt.xlabel("Sub Category")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

4️⃣ Customer Demographics

```python

bins = [0,25,35,45,55,100]
labels = ["18-25","26-35","36-45","46-55","55+"]
df["Age Group"] = pd.cut(df["Customer Age"], bins=bins, labels=labels)
age_sales = df.groupby("Age Group", observed=True)["Quantity"].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(
    data=age_sales,
    x="Age Group",
    y="Quantity",
    hue="Age Group",
    legend=False,
    width=0.4
)
plt.title("Purchases by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Purchases")
plt.show()
```

5️⃣ Location Based Sales 🌍

```python
country_sales = df.groupby("Country", observed=True)["Revenue"].sum().reset_index()
country_sales = country_sales.sort_values("Revenue", ascending=False)

plt.figure(figsize=(10,5))

sns.barplot(
    data=country_sales,
    x="Country",
    y="Revenue",
    hue="Country",      # add hue
    palette="viridis",
    width=0.4,
    legend=False        # hide legend
)

plt.title("Revenue by Country")
plt.xlabel("Country")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)

plt.show()
```

6️⃣ Profit Analysis 💰

```python
df["Profit"] = df["Revenue"] - df["Cost"]
top_profit = df.groupby("Sub Category", observed=True)["Profit"].sum().reset_index()
top_profit = top_profit.sort_values("Profit", ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(
    data=top_profit,
    y="Sub Category",
    x="Profit",
    hue="Sub Category",   # hue assign karo taake palette warning na aaye
    palette="Greens",     # Seaborn automatically bars ko alag shade deta hai
    dodge=False            # same bar me hue dublicate na ho
)

# Titles & labels
plt.title("Top 10 Profitable Sub Categories", fontsize=16)
plt.xlabel("Profit", fontsize=12)
plt.ylabel("Sub Category", fontsize=12)

# Bar value labels (optional, professional look ke liye)
for index, row in top_profit.iterrows():
    plt.text(row["Profit"] + 0.01*top_profit["Profit"].max(), index, f'{row["Profit"]:.0f}', va='center')

plt.tight_layout()
plt.legend([],[], frameon=False)  # Legend remove kar do
plt.show()
```

```python
df["Profit Margin"] = (df["Profit"] / df["Revenue"]) * 100
profit_category = df.groupby("Product Category", observed=True)["Profit Margin"].mean()
print(profit_category)
```

```python
# Grouping by product category and calculating mean profit margin
profit_category = df.groupby("Product Category", observed=True)["Profit Margin"].mean().reset_index()
plt.figure(figsize=(4,4))
plt.pie(
    profit_category["Profit Margin"], 
    labels=profit_category["Product Category"], 
    autopct="%1.1f%%",
    startangle=150,
    shadow=True
)
plt.title("Profit Margin by Product Category")
plt.show()
```
