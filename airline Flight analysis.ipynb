Q.9. What will be the Average Price of Vistara airline for a flight from Delhi to Hyderabad in Business Class ?

```python
vistara = df[
    (df["airline"] == "Vistara") &
    (df["source_city"] == "Delhi") &
    (df["destination_city"] == "Hyderabad") &
    (df["class"] == "Business")
]
vistara
```

```python
avg_price = vistara['price'].mean()
print(f"Average Price (Vistara | Delhi â†’ Hyderabad | Business): â‚¹ {avg_price:.2f}")
```

```python
plt.figure(figsize=(5,3))
sns.barplot(x="airline", y="price", data=vistara, color="purple",width=0.1)
plt.title("Average Price of Vistara Airline (Delhi â†’ Hyderabad | Business)", fontsize=12, pad=10)
plt.xlabel("Airline", fontsize=12)
plt.ylabel("Average Price (PKR)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
